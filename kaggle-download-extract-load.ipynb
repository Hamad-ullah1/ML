```python
#If you're using Kaggle datasets with opendatasets, you can do:
import kagglehub
import pandas as pd
import os

# Download dataset
path = kagglehub.dataset_download("rakeshrau/social-network-ads")

# Load CSV
df = pd.read_csv(os.path.join(path, "Social_Network_Ads.csv"))

# Save in current directory
df.to_csv("Social_Network_Ads.csv", index=False)

print("Saved successfully!")
```

Open a ZIP file as if it were a CSV/text file.

```python
#Check what's inside the ZIP
import zipfile

with zipfile.ZipFile("recipe-ingredients-dataset.zip", "r") as z:
    print(z.namelist())
```

```python
#Extract the ZIP
with zipfile.ZipFile("recipe-ingredients-dataset.zip", "r") as z:
    z.extractall("recipe_dataset")
```
